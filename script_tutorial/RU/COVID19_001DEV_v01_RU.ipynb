Aнализ данных в R

Часть 1 - Подготовка среды
--------------------
Важным элементом программирования является подготовка среды для анализа. В случае R следует отметить следующие возможности:

1. Местная среда ограниченная существующими способами:
    1. С использованием редакторов т. к.:  [RStudio](https://rstudio.com/#RStudio)
    2. С использованием редакторов которые позволяют легко выделить нужные фрагменты кода т. к.:  [Jupyter](https://jupyter.org/)
2. Решения с помощью облака которые облегчают работу:
    1. Google [Colaboratory](https://colab.research.google.com/notebooks/intro.ipynb)
    2. Amazon [SageMaker](https://aws.amazon.com/sagemaker/)
    3. Microsoft [Azure](https://azure.microsoft.com/pl-pl/)
3. Решения с помощью облака с возможностью свободной конфигурации:
    1. Amazon [EC2](https://aws.amazon.com/ec2/)

Вышеупомянутых разрешений существует гораздо больше, эти были упомянуты в качестве примеров. <br><br>
Для целей курса были использованы Jupyter [Jupyter](https://jupyter.org/), в среде [Anaconda](https://www.anaconda.com/), которая облегчает использование нужных инструментов. Скрипты можно запускать в любой среде для целей курса было взята одна среда в качестве примера, с использованием локальных инструментов. 

# 1.  Инсталляция и подготовка среды 

## 1.1	Файлы необходимые для инсталляции 

Первым этапом является загрузка среды [Anaconda](https://www.anaconda.com/), смотри секцию `Products`. Существует несколько способов, в некоммерческих целях можно выбрать версию : `Individual Edition`.


![cardinal](../_pictures/001_download.png)
<br><br>


## 1.2 Инсталляция програмной среды

После скачивания следует установить програмную среду на местном компьютере. С этой целью
нужно запустить загруженный файл и пройти процесс инсталляции, после которого можно работать с уровня консоли, либо графической накладки. В случае, если есть желание запустить редактор Jupiter в другой секции, нужно использовать консоль. 
<br>
Первым шагом является создание виртуальной среды для языка R. В примере было использовано название `renv`.
![cardinal](../_pictures/002_renv_install.png)

## 1.3 Запуск среды

Первым шагом является активация ранее созданной виртуальной среды для языка R. Как правило, работа в таковой среде ведётся в языке Python, с линейкой  `root`. Этот шаг выполняется в секции `Environment`.

![cardinal](../_pictures/003_renv_activate.png)
<br>
Далее нужно запустить редактор Jupiter, в котором можем также открыть ранее загруженные руководства для языка Jupiter. Файлы находятся в секции `Pobierz`, где также можно найти файлы для репозитория [GitHub](https://github.com/).

![cardinal](../_pictures/004_jupyter_launch.png)

<br>

После клонирования репозитория, либо загрузки интересующего содержания (скрипты и дополнительные файлы упоминаются в каждым скрипте в секции с декларируемыми переменными), можно запустить только ту часть, которая нас интересует. 

![cardinal](../_pictures/005_uruchomienie_skryptu.png)
<br><br>



## 1.4	Запуск кода

В качестве примера было выбрано код под названием `COVID19_001ETL_v23_PL.ipynb`. На данном этапе следует обозначить, что расширение `*ipybn` относится к файлам созданным с помощью редактора Jupiter.
<br>
С целью запуска данного фрагмента можно использовать комбинацию клавиш `ctrl+Enter`, либо использовать линейку `Kernel -> Run Cell`, можно также запустить код целиком: `Kernel -> Run All`.
